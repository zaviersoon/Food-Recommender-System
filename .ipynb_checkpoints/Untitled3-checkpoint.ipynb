```python
term = ['asian fusion', 'halal', 'indian', 'thai', 'vegan', 'vegetarian', 'vietnamese']
offset = np.arange(0,800,50)
tuples = list(product(term, offset))
```

```python
listings = []

cols = ['id','name', 'is_closed','review', 'rating', 'address1', 'address2', 'address3', 'zip_code', 'country', 'lat', 'long']

for term, step in tuples:
    parameters = {
        'location' : 'Singapore',
        'term' : term,
        'limit' : 50,
        'radius' : 40000,
        'offset' : step,
        'coordinates': {"latitude": 1.357107, "longitude": 103.8194992}
    }
    response = r.get(url, headers = headers, params = parameters)
    data = response.json()
    for business in data['businesses']:
        ids = business['id']
        name = business['name']
        is_closed = business['is_closed']
        reviews = business['review_count']
        rating = business['rating']
        address1 = business['location']['address1']
        address2 = business['location']['address2']
        address3 = business['location']['address3']
        zip_code = business['location']['zip_code']
        country = business['location']['country']
        lat = business['coordinates']['latitude']
        long = business['coordinates']['longitude']
        listings.append([ids, name, is_closed, reviews, rating, address1, address2, address3, zip_code, country , lat, long])
        df1 = pd.DataFrame.from_records(listings, columns = cols)
    time.sleep(3)
df1.head()
```
