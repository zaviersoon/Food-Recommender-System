#### c) Sentiment Analysis of User Reviews

<b> Define Functions </b>

```python
## Sentiment Analysis for Reviews

def plot_top_words(df, col, n, n_gram_range, title, palette, stop_words):
    
    def get_top_n_words(corpus, n = n, k = n_gram_range):     
        cv = CountVectorizer(ngram_range=(k,k), stop_words = stop_words)   
        bag_of_words = cv.fit_transform(corpus)     
        sum_words = bag_of_words.sum(axis=0)      
        words_freq = [(word, sum_words[0, idx]) for word, idx in cv.vocabulary_.items()]    
        words_freq = sorted(words_freq, key = lambda x: x[1], reverse=True) 
        return words_freq[:n]
    
    temp_df = pd.DataFrame(data=get_top_n_words(df[col], n), columns=['word','freq'])
    plt.figure(figsize=(10,6))
    sns.barplot(data=temp_df, y='word', x='freq', palette=palette)
    plt.ylabel('Words')
    plt.xlabel('Frequency')
    plt.title(title, fontsize = 15)
    plt.tight_layout()
```

```python
# Define a function to plot word cloud

def plot_wordcloud(df, col, title, background_color, colormap):
    wordcloud = WordCloud(max_words= 100,
                          width = 800, height = 800,
                          background_color = background_color,
                          colormap = colormap, 
                          random_state = 42,
                          stopwords = stopwords,
                          collocations = False,
                          contour_width = 4,
                          min_font_size = 5).generate(' '.join(df[col]))   

    # plot the WordCloud image                      
    plt.figure(figsize = (7, 7), facecolor = None)
    plt.imshow(wordcloud, interpolation = 'bilinear')
    plt.title(title, fontsize = 16, y = 1.01)
    plt.axis("off")
    plt.tight_layout()
    plt.show()
```

```python
# Print English stopwords

stopwords = stopwords.words('english')
stopwords.extend(['food', 'good', 'like', 'place', 'restaurant', 'use', 'would'])

print(stopwords)
```

<b> Vader Sentiment Analysis </b>

```python
# Initialise VADER
sent_analyzer = SentimentIntensityAnalyzer()

# Analyzing sentiment with VADER
sentiment = reviews['comment_text'].apply(lambda x : sent_analyzer.polarity_scores(x))
reviews = pd.concat([reviews, sentiment.apply(pd.Series)], axis = 1)
reviews.reset_index(drop = True, inplace = True)
reviews['compute_score'] = reviews['compound'].apply(lambda c: 'pos' if c >=0 else 'neg')
```

```python
# Sentiment Analysis of User Reviews

comp_score = reviews['compound'].value_counts().sort_index(ascending=False)

trace = go.Histogram(x = comp_score.index,
               y = comp_score.values
               )

# Create layout
layout = dict(title = 'Sentiment Analysis of User Reviews (Average Score:{:.3f})'.format(reviews['compound'].mean()),
              autosize = False, width = 800, height = 600, 
              margin=dict(l=50,r=50,b=100,t=100,pad=4),
              xaxis = dict(title = 'Compound Score', dtick = 0.1),
              yaxis = dict(title = 'Count'),
              shapes= [{'line': {'color': '#FDAB5A', 'dash': 'solid', 'width': 2},
                        'type': 'line', 'x0': reviews['compound'].mean(), 'x1': reviews['compound'].mean(),
                        'xref': 'x', 'y0': -0.1, 'y1': 600}])

# Create distribution plot

fig = go.Figure(data=[trace], layout=layout)
iplot(fig);
```
