## 3. Location-Based Recommender

```python
# Putting the Coordinates of Restaurants together into a dataframe

coordinates = central[['lat','long']]
coordinates
```

```python
# Elbow method to determine the number of K in Kmeans Clustering

def km_elbow_plot(cluster_num, df):
    k_list = range(2, cluster_num)# based on prior runs sil is 0.5 at 19, 0.7 at 49
    X = df[['lat', 'long']]
    Sum_of_squared_distances = []
    sil_score = []
    for k in k_list:
        km = KMeans(n_clusters = k, random_state = 42)
        km = km.fit(X)
        Sum_of_squared_distances.append(km.inertia_)
        sil_score.append(silhouette_score(X, km.labels_))
    plt.plot(k_list, Sum_of_squared_distances, 'bx-')
    plt.xlabel('k')
    plt.ylabel('Sum_of_squared_distances')
    plt.title('SSE For by k')
    plt.show()

    plt.plot(k_list, sil_score, 'bx-')
    plt.xlabel('k')
    plt.ylabel('Silhouette Score')
    plt.title('Silhouette Score by k')
    plt.show()
```

```python
# Elbow Graph

km_elbow_plot(50, coordinates)
```

```python
# Let's try k = 3

km = KMeans(n_clusters = 3, random_state = 42)
km.fit(coordinates)
print("k = 3", " silhouette_score ", silhouette_score(coordinates, km.labels_, metric='euclidean'))
```

```python
# Predict all the restraurant cluster

central['cluster'] = km.predict(central[['lat','long']])
central.head()
```

## 4. Recommendation

```python
# Define function to recommend restaurant based on location 

def location_based_recommendation(df, latitude, longitude):
    
    """Predict the cluster for longitude and latitude provided"""
    cluster = km.predict(np.array([latitude, longitude]).reshape(1,-1))[0]
    

    """Get the best restaurant in this cluster along with the relevant information for a user to make a decision"""
    return central[central['cluster'] == cluster].sort_values(by=['restaurants_rating', 'review'], ascending=False).iloc[0:10]
```
